# Additional Columns
## 1. Weekend

```python
df['Start Time'] = pd.to_datetime(df['Start Time'])
df['Weekend'] = (df['Start Time'].dt.dayofweek > 4)

# check if there are trips where a bike rental started on a weekday
# and was ended on a weekend or vice versa
df['End Time'] = pd.to_datetime(df['End Time'])
df['Weekend_End'] = (df['End Time'].dt.dayofweek > 4)
df
```

```python
# Check if Weekend Start & Weekend End values are unequal in some rows
df[df['Weekend'] != df['Weekend_End']]
```
