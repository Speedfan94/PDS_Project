```python
import pandas as pd
```

```python
df = pd.DataFrame({'Col1': [10, 20, 16, 15, 30, 35, 45],
                   'Trip': ['start', 'end', 'end', 'start', 'start', 'start', 'end'],
                   'Col3': [17, 27, 12, 22, 37, 40, 52]})
df
```

```python
sr_noise = df['Trip'] != df['Trip'].shift()
sr_noise
df['valid_start'] = sr_noise
sr_noise_end = df['Trip'] != df['Trip'].shift(-1)
sr_noise_end
df['valid_end'] = sr_noise_end
df['valid'] = ((df['Trip'] == 'start') & (df['valid_start'] == True)) | ((df['Trip'] == 'end') & (df['valid_end'] == True))
df
```

```python
df.sort_values(["Trip", "valid_start"])
df
```
