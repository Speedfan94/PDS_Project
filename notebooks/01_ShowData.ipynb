### Drop First and Last from here

```python
#only look at start and end
df_clean_no_duplicates_without_first_last=df_clean_no_duplicates[(df_clean_no_duplicates["trip"]=="start") | (df_clean_no_duplicates["trip"]=="end")]
#example bike
df_clean_no_duplicates_without_first_last[df_clean_no_duplicates_without_first_last["b_number"]==11281].head(30)
```

#### Split Start / End

```python
#Create Df for Start
df_clean_start=df_clean_no_duplicates_without_first_last[df_clean_no_duplicates_without_first_last["trip"]=="start"
                                                        ].rename(columns={"datetime":"Start Time", 
                                                                          "b_number":"Bike Number", 
                                                                          "Latitude":"Start Lat", 
                                                                          "Longitude":"Start Long"})
df_clean_start.drop("trip", axis=1, inplace=True, errors="ignore")

#Create Df for End
df_clean_end=df_clean_no_duplicates_without_first_last[df_clean_no_duplicates_without_first_last["trip"]=="end"
                                                        ].rename(columns={"datetime":"End Time", 
                                                                          "b_number":"Bike Number", 
                                                                          "Latitude":"End Lat", 
                                                                          "Longitude":"End Long"})
df_clean_end.drop("trip", axis=1, inplace=True, errors="ignore")
```

```python
#Compare number of Startpoints and Endpoints for each Bike

sr_starts_each_bike=df_clean_start.groupby(by="Bike Number").count()["Start Time"]

sr_ends_each_bike=df_clean_end.groupby(by="Bike Number").count()["End Time"]

df_starts_ends_each_bike = pd.DataFrame(data={"Starts":sr_starts_each_bike, "Ends":sr_ends_each_bike})
df_starts_ends_each_bike["Difference"]=sr_starts_each_bike-sr_ends_each_bike

df_starts_ends_each_bike
#Not the same amount of start and end points => Problems in merging Start and End of a trip
```

```python
#How many Bikes differ in number of start and endpoints
num_diff=df_starts_ends_each_bike[df_starts_ends_each_bike["Difference"]>0].index.size
num_equ=df_starts_ends_each_bike[df_starts_ends_each_bike["Difference"]==0].index.size

print("There are", num_diff, "Bikes with more starts than ends\nOnly",
      num_equ, "bikes do have the same amaount on start and end points")
```
